```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
tips = pd.read_csv('C:\\Users\\SAHITHYA AAKULA\\Downloads\\tips.csv')
```

```python
# Unique values in the 'day' column
unique_days = tips['day'].unique()
# Print the unique values
print(unique_days)
```

```python
plt.figure(figsize=(12, 8))
```

```python
sns.displot(tips['total_bill'],kde = False,bins=50)
```

```python
sns.jointplot(x=tips['total_bill'],y='tip',data=tips, kind='hex')
sns.jointplot(x=tips['total_bill'],y='tip',data=tips, kind='reg')
```

```python
# Scatter plot Total Bill vs Tip with day color coding
plt.subplot(2, 2, 1)
sns.scatterplot(data=tips, x='total_bill', y='tip', hue='day')
plt.title('Total Bill vs Tip by Day')
```

```python
# Scatter plot Total Bill vs Tip with time color coding
plt.subplot(2, 2, 2)
sns.scatterplot(data=tips, x='total_bill', y='tip', hue='time')
plt.title('Total Bill vs Tip by Time')
```

```python
# Scatter plot Total Bill vs Tip with smoker status
plt.subplot(2, 2, 3)
sns.scatterplot(data=tips, x='total_bill', y='tip', hue='smoker')
plt.title('Total Bill vs Tip by Smoker Status')
```

```python
# Scatter plot Total Bill vs Tip with size
plt.subplot(2, 2, 4)
sns.scatterplot(data=tips, x='total_bill', y='tip', hue='size')
plt.title('Total Bill vs Tip by Party Size')
```

```python
sns.pairplot(tips)
sns.pairplot(tips,hue='sex')
```

```python
sns.barplot(x='sex',y='total_bill',data=tips)
import numpy as np
sns.barplot(x='sex',y='total_bill',data=tips,estimator=np.std)
```

```python
# Bar plot Average tip by time
plt.subplot(2, 2, 2)
sns.barplot(x='time', y='tip', data=tips)
plt.title('Average Tips by Time')
```

```python
# Bar plot Average tip by smoker status
plt.subplot(2, 2, 3)
sns.barplot(x='smoker', y='tip', data=tips)
plt.title('Average Tips by Smoker Status')
```

```python
# split violin plot showing day and time interaction
plt.figure(figsize=(12, 6))
sns.violinplot(x='day', y='tip', hue='time', split=True, data=tips)
plt.title('Tips Distribution by Day and Time (Split)')
plt.show()
```

```python
numeric_tips = tips.select_dtypes(include=[float,int])
tc = numeric_tips.corr()
```
